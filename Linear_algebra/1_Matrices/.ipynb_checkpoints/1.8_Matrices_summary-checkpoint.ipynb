# Matrices summary


:::{list-table}
:header-rows: 1

* - Operation
  - <p style="text-align:center"> Definition </p>
  - <p style="text-align:center"> Example </p>
* - $m \times n $ matrix
  - \begin{align*} 
        \begin{pmatrix} 
            a_{11} & a_{12} & \cdots & a_{1m} \\  
            a_{21} & a_{22} & \cdots & a_{2m} \\ 
            \vdots & \vdots & \ddots & \vdots \\ 
            a_{m1} & a_{m2} & \cdots & a_{mn} 
        \end{pmatrix} 
    \end{align*}
  - \begin{align*} \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \end{align*} 
* - Matrix indexing 
  - $$[A]_{ij} = a_{ij}$$
  - $$a_{21} = 3$$
* - Addition of two matrices
  - $$[A + B]_{ij} = [A]_{ij} + [B]_{ij}$$
  - \begin{align*} 
      \phantom{=} &\begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} + 
      \begin{pmatrix} 1 & -2 \\ 4 & 1 \end{pmatrix}\\
      =& \begin{pmatrix} 2 & 0 \\ 7 & 5 \end{pmatrix} 
  \end{align*}
* - Multiplication  of a matrix by a scalar
  - $$k[A]_{ij} = [kA]_{ij}$$
  - \begin{align*} 2 
      \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} = 
      \begin{pmatrix} 2 & 4 \\ 6 & 8 \end{pmatrix} 
  \end{align*}
* - Multiplication of two matrices
  - $$[AB]_{ij} = \displaystyle\sum_{k=1}^n [A]_{ik}[B]_{jk} $$ 
  - \begin{align*}
        \phantom{=} &\begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} 
        \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} \\
        = &\begin{pmatrix} 5 + 14 & 6 + 16 \\ 15 + 28 & 18 + 32 \end{pmatrix} \\
        = &\begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
    \end{align*}
* - Diagonal matrix 
  - $$[D]_{ij} = 0 \quad \forall i \neq j $$ 
  - \begin{align*} 
        \begin{pmatrix} 
          1 & 0 & 0 \\ 
          0 & 2 & 0 \\ 
          0 & 0 & 3
        \end{pmatrix}
    \end{align*}
* - Zero matrix
  - $$[\mathbf{0}]_{ij} = 0 \quad \forall i, j $$
  - \begin{align*} 
        \begin{pmatrix} 
          0 & 0 & 0 \\ 
          0 & 0  & 0 
        \end{pmatrix}
    \end{align*}
* - Identity matrix
  - $$[I]_{ij} = \begin{cases} 1, & i == j, \\ 0, & i \neq j \end{cases}$$
  - $$\begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}$$
:::


- [Symmetric matrices](symmetric-matrix-definition) are square matrices where the value off the main diagonal are symmetric about the main diagonal.
- The [determinant](determinant-section) of a square matrix is a scalar value.
- The [determinant of a $2\times 2$ matrix](2x2-determinant-definition) is calculated as the product of the elements on the main diagonal minus the product of the other two elements.
- The [determinant of an $n\times n$ matrix](2x2-determinant-definition) is calculated by multiplying the elements along a row or column of the matrix by their [co-factors](cofactor-definition) and summing the result. 
- The [adjoint](adjoint-definition) of a square matrix is calculated by transposing the matrix of [co-factors](cofactor-definition).
- The [inverse of a square matrix](inverse-matrix-definition) is the multiplicative inverse of the matrix. 
- Matrices with a determinant of zero do not have an inverse and are called **singular** matrices.
- The inverse of a matrix can be calculated using the [adjoint-determinant formula](adjoint-determinant-formula-definition).
- Equations involving matrices and by manipulated using [matrix algebra](matrix-algebra-section) using addition, subtraction, multiplication by a scalar and matrix multiplication operations.
